```python
import safep
sa = safep.StructuralAnalysis()
sa.make_node(0.0,0.0)
sa.make_node(1.0,0.0)
sa.make_node(2.0,0.0)
sa.make_node(0.0,1.0)
sa.make_node(1.0,1.0)
sa.make_truss_member(0,1)
sa.make_truss_member(1,2)
sa.make_truss_member(3,4)
sa.make_truss_member(0,4)
sa.make_truss_member(1,4)
sa.make_truss_member(2,4)

sa.set_boundary_condition(0,x_fix=True,y_fix=True)
sa.set_boundary_condition(3,x_fix=True,y_fix=True)
sa.set_boundary_condition(1,y_fix=True)

sa.apply_load(2,y_force=-1.0)
#sa.apply_temperature(2,1.0,T0=1.0)

#sa.draw_structure()
sa.solve()
sa.draw_structure()
#sa.draw_deformation(def_scale=0.1)
sa.draw_axial_force()
sa.draw_strech()
sa.print_result()
#sa.print_element()
#sa.printnow()
```

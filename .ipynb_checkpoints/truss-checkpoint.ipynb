```python
import safep
sa = safep.StructuralAnalysis()
sa.make_node(0.0,0.0)
sa.make_node(1.0,0.0)
sa.make_node(2.0,0.0)
sa.make_node(1.0,-1.0)
sa.make_truss_member(0,3)
sa.make_truss_member(1,3)
sa.make_truss_member(2,3)

sa.set_boundary_condition(0,x_fix=True,y_fix=True)
sa.set_boundary_condition(1,x_fix=True,y_fix=True)
sa.set_boundary_condition(2,x_fix=True,y_fix=True)

#sa.apply_load(2,x_force=-1.0)
sa.apply_temperature(2,1.0,T0=1.0)

#sa.draw_structure()
sa.solve()
sa.draw_structure()
#sa.draw_deformation(def_scale=0.1)
sa.draw_axial_force()
sa.draw_strech()
sa.print_result()
#sa.print_element()
#sa.printnow()
```
